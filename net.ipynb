```julia
"""
Lexer for additional data on nodes in a nexus tree
"""
function lnodedata(s)
    d = Dict()
    i = 1

    while true
        m = match(r"^(.*?)\=", s[i:end]) ## Everything until the equal sign
        if m == nothing
            break
        end
        key = m.captures[1]
        i += length(key) + 1

        if s[i] == '{'
            val = match(r"\{(.*?)\}", s[i:end]).match ## First match inbetween curly brackets
        else
            val = match(r"^(.*?),", s[i:end]).captures[1] ## Everything until the comma
        end

        i += length(val) + 1
        d[key] = val
    end
    d
end
```
